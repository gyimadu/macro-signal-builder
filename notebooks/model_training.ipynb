```python
import pandas as pd

df = pd.read_csv("../data/final_dataset.csv")

df['Date'] = pd.to_datetime(df['Date'])

df.set_index('Date', inplace=True)

df.head()
```

```python
assets = ['spy', 'qqq', 'tlt', 'gld', 'hyg', 'uup']

for asset in assets:
    df[f'{asset}_target'] = (df[f'{asset}_returns'].shift(-1) > 0).astype(int)

df
```

```python
df.info()
```

```python
target = 'spy_target'

features = [
    'gdp', 'gdp_growth', 'cpi', 'unemployment', 'fedFunds',
    'spy_volatility', 'qqq_volatility', 'tlt_volatility', 'gld_volatility', 'hyg_volatility', 'uup_volatility',
    'qqq_returns', 'tlt_returns', 'gld_returns', 'hyg_returns', 'uup_returns',
]

X = df[features]
y = df[target]
```

```python
n = len(df)
train_end = int(n*0.6)
val_end = int(n*0.8)

X_train = X.iloc[:train_end]
y_train = y.iloc[:train_end]

X_val = X.iloc[train_end:val_end]
y_val = y.iloc[train_end:val_end]

X_test = X.iloc[val_end:]
y_test = y.iloc[val_end:]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)

# model initialization for spy_target prediction
spy_model = LogisticRegression(max_iter=1000)
#train on training set
spy_model.fit(X_train_scaled, y_train)
# do first evaluation on validation set. NB: testing set is not used for first evaluation to reduce risk of overfitting
y_val_pred = spy_model.predict(X_val_scaled)
```

```python
# evaluation of performance on validation set
from sklearn.metrics import confusion_matrix
accuracy = accuracy_score(y_val, y_val_pred)

print(f"val accuracy: {accuracy}")
print(f"{classification_report(y_val, y_val_pred)}")

conf_matrix = confusion_matrix(y_val, y_val_pred)

print(conf_matrix)



```

```python
# from the model evaluation, model accurately predicts all positive cases but misses all negative cases
# this is because the dataset is imbalanced, with more positive cases (31) than negative cases (11)
# to address this, I use a balanced class weight to give equal importance to both classes
spy_model_balanced = LogisticRegression(max_iter=1000, class_weight='balanced')
spy_model_balanced.fit(X_train_scaled, y_train)
```

```python
y_val_pred_balanced = spy_model_balanced.predict(X_val_scaled)

print(f"val accuracy: {accuracy_score(y_val, y_val_pred_balanced)}")
print(f"{classification_report(y_val, y_val_pred_balanced)}")

conf_matrix_balanced = confusion_matrix(y_val, y_val_pred_balanced)

print(conf_matrix_balanced)
```

```python
# from the value counts for target variables in the training wet, the class 0 is underrrepresented which could the reason balanced class weights my not be helping
# also, logistic regression may not be capturing nonlinear feature interactions that distinguish the class 0 from the class 1
# I will try a random forest model to see if it can capture these interactions

from sklearn.ensemble import RandomForestClassifier

spy_model_rf = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)
spy_model_rf.fit(X_train_scaled, y_train)
```

```python
y_val_pred_rf = spy_model_rf.predict(X_val_scaled)

print(f"val accuracy: {accuracy_score(y_val, y_val_pred_rf)}")
print("\nclassification report\n", classification_report(y_val, y_val_pred_rf))
print("\nconfusion matrix:\n", confusion_matrix(y_val, y_val_pred_rf))
```
